```python

from ultralytics import YOLO
import matplotlib.pyplot as plt
import os

# Load the trained YOLO model
#model_path = "/content/yolov8n.pt"  # Replace with your model path
#model = YOLO(model_path)

# Test with a random image (update path as needed)
image_path = "/content/gdrive/MyDrive/Vehicles/test/Big Truck/Image_000005.jpg"  # Replace with an image unrelated to cats/dogs

# Run inference with a confidence threshold
results = model(image_path, conf=0.5)  # Adjust confidence as needed

# Filter results to only include "cats" and "dogs" (class IDs: 0 and 1)
target_classes = [0, 1]
detections = [
    box for box in results[0].boxes.data
    if int(box[-1]) in target_classes and box[-2] > 0.9  # Confidence threshold
]

# Display results if detections exist
if detections:
    print("Detected objects belonging to target classes:")
    for det in detections:
        print(f"Class: {int(det[-1])}, Confidence: {det[-2]:.2f}, BBox: {det[:4]}")
    annotated_image = results[0].plot()
    plt.imshow(annotated_image)
    plt.axis('off')
    plt.title("Detection Results")
    plt.show()
else:
    print("No valid detections for the target classes in this image.")
```

```python
import os
from sklearn.metrics import accuracy_score

# Folder containing "Van" images
van_folder = "/content/gdrive/MyDrive/Vehicles/test/Van"

# Run inference on all images in the folder
true_labels = []
predicted_labels = []

# Loop through all images in the folder
for image_name in os.listdir(van_folder):
    image_path = os.path.join(van_folder, image_name)
    if os.path.isfile(image_path):
        # Run inference
        results = model(image_path, conf=0.9)  # Adjust confidence as needed

        # Filter results for target classes (including "Van" class 6)
        target_classes = [0, 1, 2, 3, 4, 5, 6]
        detections = [
            box for box in results[0].boxes.data
            if int(box[-1]) in target_classes and box[-2] > 0.9  # Confidence threshold
        ]

        # Assign predicted label as "6" if any detection matches the "Van" class
        if any(int(box[-1]) == 6 for box in detections):
            predicted_labels.append(6)
        else:
            predicted_labels.append(-1)  # Assume -1 for no valid prediction

        # True label for all images in the "Van" folder
        true_labels.append(6)

# Calculate accuracy
if true_labels and predicted_labels:
    accuracy = accuracy_score(true_labels, predicted_labels)
    print(f"Accuracy of detecting 'Van' class: {accuracy * 100:.2f}%")
else:
    print("No valid predictions to calculate accuracy.")
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the pre-trained YOLO model
#model = YOLO('yolov8n.pt')  # Use yolov8n.pt for the nano model; you can choose other weights.

# Path to the image
image_path = '/content/gdrive/MyDrive/RANDOM INTERNET IMAGES/RANDOM INTERNET IMAGES/'

# Load the image
image = cv2.imread(image_path)

# Perform object detection
results = model.predict(source=image_path, save=False, conf=0.5)  # Set a confidence threshold

# Count the detected objects
object_counts = {}
for result in results:
    for cls in result.boxes.cls:
        class_name = model.names[int(cls)]  # Map class ID to class name
        object_counts[class_name] = object_counts.get(class_name, 0) + 1

# Print object counts
print("Object Counts:")
for obj, count in object_counts.items():
    print(f"{obj}: {count}")

# Optional: Annotate the image with bounding boxes
annotated_image = results[0].plot()  # Get the first image with annotations

# Display the image
plt.imshow(cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# Save the annotated image (optional)
cv2.imwrite('annotated_image.jpg', annotated_image)
```
